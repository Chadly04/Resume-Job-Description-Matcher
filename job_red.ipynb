This is the input of Resume and Job description

```python
resume = """ Your resume content here
"""




job = """ Your job description content here
"""
```

Imports for cleaning 

```python
import re 
from nltk.corpus import stopwords
```

```python
stop =  stopwords.words("english")
```

cleaning the data 

```python
def clean(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9+#./'\-\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    tokens = text.split()
    cleaned_tokens = []
    for word in tokens:
        if word not in stop:
            cleaned_tokens.append(word)
    return cleaned_tokens
c_res = clean(resume)
c_job = clean(job)
```

Joining the data in a list 

```python
clean_resume_text = " ".join(c_res)
clean_job_text = " ".join(c_job)
print(clean_job_text)
```

Using google Gemini for extracting keywords
%pip install -U google-genai # run command so you can use genai

```python
import os

from google import genai

GOOGLE_API_KEY = os.getenv("Google_api") # need the get os for api from system

client = genai.Client(api_key=GOOGLE_API_KEY)

def extract_keywords (text):
   prompt = f"""
   Extract the most important resume/job-matching keywords from the text below.

    Focus on these categories:
    - programming languages
    - frameworks and libraries
    - databases
    - tools and platforms
    - technical skills
    - soft skills
    - education requirements
    - experience level

    Return only a comma-separated list.
    Do not write full sentences.

    Text:
    {text}
    """
   response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
   return response.text
```

Prints results from the genai calls 

```python
res_keyWord = extract_keywords(resume)
job_keyWord = extract_keywords(job)

print(res_keyWord)
print(job_keyWord)  
```

Turns them back to a list 

```python
res_keyWord = [word.strip().lower() for word in res_keyWord.split(',') if word.strip()]
job_keyWord = [word.strip().lower() for word in job_keyWord.split(',') if word.strip()]
print(f"Resume Keywords: {res_keyWord}")
```

```python
print(f"Resume Keywords: {res_keyWord}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Comparte Key words using TF-IDF & cosine similarity

```python
documents = [" ".join(res_keyWord), " ".join(job_keyWord)] ## Convert the keyword lists to strings``

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
match_score = similarity * 100

print("TF-IDF Cosine Similarity:", similarity)
print("Match Score:", round(match_score, 2), "%")
```

Comparte cleaned text words using TF-IDF & cosine similarity

```python
documents = [clean_resume_text, clean_job_text] ## Convert the keyword lists to strings``

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
match_score = similarity * 100

print("TF-IDF Cosine Similarity:", similarity)
print("Match Score:", round(match_score, 2), "%")
```

Compare key_words  using Bag of words and cosSim 

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [" ".join(res_keyWord), " ".join(job_keyWord)]
vectorizer = CountVectorizer()  
count_matrix = vectorizer.fit_transform(documents)  

similarity = cosine_similarity(count_matrix[0:1], count_matrix[1:2])[0][0]
match_score = similarity * 100

print(f"Match Score: {match_score:.2f}%")

```

Compare cleaned text using Bag of words and cosSim 

```python
documents = [clean_resume_text, clean_job_text]
vectorizer = CountVectorizer()  
count_matrix = vectorizer.fit_transform(documents)  

similarity = cosine_similarity(count_matrix[0:1], count_matrix[1:2])[0][0]
match_score = similarity * 100

print(f"Match Score: {match_score:.2f}%")
```
